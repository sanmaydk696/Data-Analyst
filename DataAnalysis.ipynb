Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
import pandas as pd
import numpy as np
from scipy import stats

# Assuming you have a DataFrame named 'df' with a 'ConvertedComp' column
z_scores = np.abs(stats.zscore(df['ConvertedComp']))
threshold = 3  # Adjust this threshold as needed

# Filter rows where Z-score is within the threshold
df_no_outliers = df[(z_scores < threshold)]

# Now 'df_no_outliers' contains the DataFrame with outliers removed
df_no_outliers.median()
```
